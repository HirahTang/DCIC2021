```python
def area_plot(rand):
    
    x = np.array([rand['1_LATITUDE'], rand['2_LATITUDE'], rand['3_LATITUDE'], rand['4_LATITUDE'], rand['5_LATITUDE']])
    y = np.array([rand['1_LONGITUDE'], rand['2_LONGITUDE'], rand['3_LONGITUDE'], rand['4_LONGITUDE'], rand['5_LONGITUDE']])
    
    return x, y
```

```python
data = gxdc_stances[:70]
```

```python
plt.figure(figsize = (20, 10))
for index, rows in data.iterrows():
    
    x,y = area_plot(rows)
    for i in range(0, len(x)):
        plt.plot(x[i:i+2], y[i:i+2], 'ro-', markersize = 1)
plt.show()
```
